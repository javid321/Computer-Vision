```python
import matplotlib.pyplot as plt
import matplotlib as matp

from sklearn.cluster import KMeans
```

```python
image = matp.image.imread('../dataset/kucing/oyen/oyen_21.jpg')
```

```python
plt.imshow(image)
```

```python
image.shape
```

```python
image_re = image.reshape(-1, 3)
```

```python
image_re.shape
```

```python
model_kmeans = KMeans(n_clusters=10)
```

```python
model_kmeans.fit(image_re)
```

```python
segmented_img = model_kmeans.cluster_centers_[model_kmeans.labels_]
segmented_img = segmented_img.reshape(image.shape)
```

```python
plt.imshow(segmented_img/255)
```
